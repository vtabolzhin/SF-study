Написать код, который вычисляет сумму всех чисел, удовлетворяющих следующие условия:  

* положительные целые числа от 1 до 1_000_000_002 (один миллиард два) включительно  
* которые нацело (без остатка) делятся на 3 (пример: 3, 6, 9, ...)  
* и которые не заканчиваются на 4 и 7 (пример заканчивающихся на 4 и 7: 24, 27, 54, 57 ...)  

Результат суммы ввести в поле ниже.  
В поле нужно ввести только число, являющееся ответом, без пробелов и других символов.

```python
%%time

answer = 0
for i in range(1,1_000_000_003):
    if not i%3 and str(i)[-1] not in '47':
        answer += i
print(answer)
```

```python
def my_parser(input_file, output_file):
    answer = ''
    with open(input_file) as f:
            for line in f:
                line = line.split()
                result = line[1]+line[0]+line[2]
                answer += str(eval(result))+','
    answer = answer[:-1]
    new_file = open(output_file, 'w')
    new_file.write(answer)
    new_file.close()
    print('Done!')
```

```python
input_ = '003_data/test_input_file_1.txt'
output_ = '003_data/test_output_file_3.txt'
my_parser(input_, output_)
```

```python
def my_parser_2(input_file_text, input_file_indices, output_file):
    answer = ''
    with open(input_file_text) as f_txt:
        with open(input_file_indices) as f_idx:
            for text, indices in zip(f_txt, f_idx):
                indices = indices.split()
                index_low, index_high = int(indices[0]), int(indices[1])
                answer += text[index_low:index_high+1] + ' '
    with open(output_file, 'w') as f:
        f.write(answer)
    print('Done!')
```

```python
input_file_text = '003_data/test_import_file_2_1.txt'
input_file_indices = '003_data/test_import_file_2_2.txt'
output_file = '003_data/test_output_file_2_my.txt'
my_parser_2(input_file_text, input_file_indices, output_file)
```

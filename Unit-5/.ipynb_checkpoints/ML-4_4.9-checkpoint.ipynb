```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.metrics import mean_squared_error as mse

n_points = 30
x = np.linspace(0., 7., n_points)
y = np.sin(x) + np.random.rand(n_points)

y_test = np.sin(x) + np.random.rand(n_points) # генерируем данные с новым шумом

plt.figure(figsize=(15,5))
plt.scatter(x, y, color='k', label='training data')

for degree,label in zip([1,5,30], ['underfitting', 'a good fit', 'overfitted']):
    poly_coef = np.polyfit(x, y, deg=degree)
    inter_poly = np.polyval(poly_coef, x)
    plt.plot(x, inter_poly, label=label)
    
    rmse = mse(y, inter_poly)
    print(f"Degree: {degree}, rmse-train: {rmse:.2f}")
    rmse = mse(y_test, inter_poly)
    print(f"Degree: {degree}, rmse-test: {rmse:.2f}")

plt.plot(x, np.sin(x), color='r', label='sin(x)')
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.tail()
```

```python
sex_dummies = pd.get_dummies(titanic['Sex'])
titanic = titanic.drop('Sex', axis=1).join(sex_dummies)
titanic.tail()
```

```python
X = titanic.drop(['Survived','Name'], axis=1)
y = titanic['Survived']

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.15, shuffle=True, random_state=123)

clf = DecisionTreeClassifier(random_state=456)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_train)
f1_train = f1_score(y_train, y_pred)
print(f"[train] F1-score = {f1_train:.2f}")
```

```python
y_pred = clf.predict(X_test)
f1_test = f1_score(y_test, y_pred)
print(f"[test] F1-score = {f1_test:.2f}")
```

Точность при валидации значительно ниже точности при обучении. Это говорит о переобучении.  
Проверим еще раз при помощи кросс-валидации.

```python
from sklearn.model_selection import cross_validate

cv_metrics = cross_validate(clf, X, y, cv=5, scoring='f1_micro', return_train_score=True)
pd.DataFrame(cv_metrics)
```

```python
def plot_cv_metrics(cv_metrics):
    avg_f1_train, std_f1_train = np.mean(cv_metrics['train_score']), np.std(cv_metrics['train_score'])
    avg_f1_test, std_f1_test = np.mean(cv_metrics['test_score']), np.std(cv_metrics['test_score'])
    
    print(f"[train] F1-score = {avg_f1_train:.2f} +/- {std_f1_train:.2f}")
    print(f"[valid] F1-score = {avg_f1_test:.2f} +/- {std_f1_test:.2f}")
    
    plt.figure(figsize=(15,5))
    
    plt.plot(cv_metrics['train_score'], label='train', marker='.')
    plt.plot(cv_metrics['test_score'], label='valid', marker='.')
    
    plt.ylim(0, 1)
    plt.xlabel('CV-iterations', fontsize=15)
    plt.ylabel('F1-score', fontsize=15)
    plt.legend(fontsize=15)
    plt.show()
```

```python
plot_cv_metrics(cv_metrics)
```

Гипотеза о переобучении подтвердилась.  
Попробуем упростить нашу модель, а именно уменьшить глубину деревьев до 5

```python
clf = DecisionTreeClassifier(max_depth=5, random_state=456)
cv_metrics = cross_validate(clf, X, y, cv=5, scoring='f1_micro', return_train_score=True)
plot_cv_metrics(cv_metrics)
```

Целевой переменной будет 'compliance', разделите данные на обучающую и тестовую выборки в соотношении 70%/30% без перемешивания.  

Обучите DecisionTreeClassifier из scikit-learn с параметром random_state=23. Посчитайте значения метрики F1 на тренировочной выборке и на тестовой выборке. Вычтите из значения метрики на тренировочной выборки значение метрики на тестовой выборке и запишите результат в переменную result.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
vis_data = pd.read_csv("./train.csv", encoding = 'ISO-8859-1', low_memory = False)
```

```python
vis_data = vis_data[['fine_amount', 'state_fee', 'late_fee',\
                     'discount_amount', 'balance_due', 'compliance']].dropna()
X = np.array(vis_data.drop('compliance', axis=1))
y = np.array(vis_data['compliance'])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=False)
clf = DecisionTreeClassifier(random_state=23)
clf.fit(X_train, y_train)
y_pred_train, y_pred_test = clf.predict(X_train), clf.predict(X_test)
f1_train, f1_test = f1_score(y_train, y_pred_train), f1_score(y_test, y_pred_test)
result = f1_train - f1_test
print(f1_train, f1_test)
result
```

```python
from sklearn.model_selection import KFold

clf = DecisionTreeClassifier(random_state=23)
kf = KFold(n_splits=5, shuffle=True)
cv_metrics = cross_validate(clf, X, y, cv=kf, scoring='f1', return_train_score=True)
plot_cv_metrics(cv_metrics)
pd.DataFrame(cv_metrics)
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=23)
kf = KFold(n_splits=5, shuffle=True)
cv_metrics = cross_validate(clf, X, y, cv=kf, scoring='f1', return_train_score=True)
plot_cv_metrics(cv_metrics)
pd.DataFrame(cv_metrics)
```

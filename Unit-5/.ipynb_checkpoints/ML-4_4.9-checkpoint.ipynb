```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
n_points = 30
x = np.linspace(0., 7., n_points)
y = np.sin(x) + np.random.rand(n_points)

plt.figure(figsize=(15,5))
plt.scatter(x, y, color='k', label='training data')

for degree,label in zip([1,5,30], ['underfitting', 'a good fit', 'overfitted']):
    poly_coef = np.polyfit(x, y, deg=degree)
    inter_poly = np.polyval(poly_coef, x)
    plt.plot(x, inter_poly, label=label)

plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
```

```python
titanic = pd.read_csv('titanic.csv')
titanic.tail()
```

```python
sex_dummies = pd.get_dummies(titanic['Sex'])
titanic = titanic.drop('Sex', axis=1).join(sex_dummies)
titanic.tail()
```

```python
X = titanic.drop(['Survived','Name'], axis=1)
y = titanic['Survived']

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, test_size=0.15, shuffle=True, random_state=123)

clf = DecisionTreeClassifier(random_state=456)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_train)
f1_train = f1_score(y_train, y_pred)
print(f"[train] F1-score = {f1_train:.2f}")
```

```python
y_pred = clf.predict(X_test)
f1_test = f1_score(y_test, y_pred)
print(f"[test] F1-score = {f1_test:.2f}")
```

```python
def f(x, y):
    return (1 - x) ** 2 + 100 * (y - x ** 2) ** 2
def grad(x, y):
    dx = 2 * x - 2 + 100 * (-4 * y * x + 4 * x ** 3)
    dy = 100 * (2 * y - 2 * x ** 2)
    return (dx, dy)
```

```python
x0 = (0, 2)
gamma = 1e-3
x_cur = x0

vals = []
coords = []
i = 0
while True:
    x_new = (x_cur[0] - gamma * grad(*x_cur)[0],
            x_cur[1] - gamma * grad(*x_cur)[1])
    if dist(x_new, x_cur) < 1e-9:
        break
    x_cur = x_new
    vals.append(f(*x_cur))
    coords.append(x_cur)
    i += 1
    print(f"iter={i}; x=({x_cur[0]:.2f}, {x_cur[1]:.2f});"
          f" f(x)={f(*x_cur):.2f}; grad f(x)=({grad(*x_cur)[0]:.2f}, {grad(*x_cur)[1]:.2f})")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot(vals)
```

```python
n = 100 # number of discretization points along the x-axis
m = 100 # number of discretization points along the x-axis
a=-0.5; b=2. # extreme points in the x-axis
c=-1.5; d=4. # extreme points in the y-axis
X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

Z = f(X,Y)

plt.xlim(a,b)
plt.ylim(c,d)

plt.contour(X,Y,Z, np.logspace(-0.5, 3, 31),cmap='gray_r')
plt.plot([c[0] for c in coords], [c[1] for c in coords])
plt.scatter(1,1,c='red')

```

```python
import numpy as np
np.set_printoptions(suppress=True)

def f(x,y):
    return (1 - x)**2 + 100*(y - x**2)**2

def grad(x,y):
    dx = 2*(x - 1) - 400*x*(y - x**2)
    dy = 200*(y - x**2)
    return np.array([dx, dy])
```

```python
x0 = np.array([-1.5,2])
gamma = 1e-3
i = 0
x_cur = x0
vals = []
coords = []

while True:
    vals.append(f(*x_cur))
    coords.append(x_cur)
    print(f"iter = {i}, (x,y) = {x_cur}, f(x,y) = {f(*x_cur):.3f}, grad = {grad(*x_cur)}")
    if f(*x_cur) < 0.001 or i > 5000:
        break
    i += 1
    x_new = x_cur - gamma*grad(*x_cur)
    x_cur = x_new
```

```python
import matplotlib.pyplot as plt

plt.plot(vals)
plt.show()
```

```python
n = 100 # number of discretization points along the x-axis
m = 100 # number of discretization points along the x-axis
a=-2; b=1.5 # extreme points in the x-axis
c=-0.5; d=2.5 # extreme points in the y-axis
X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

Z = f(X,Y)

plt.xlim(a,b)
plt.ylim(c,d)

plt.contour(X,Y,Z, np.logspace(-0.5, 3, 31),cmap='gray_r')
plt.plot([c[0] for c in coords], [c[1] for c in coords])
plt.scatter(1,1,c='red')
plt.show()
```

```python
def f2(x,y):
    return 2*x**2 - 4*x*y + y**4 + 2

optimize.minimize(lambda x: f2(*x), (-10,10))
```

```python
def f3(w0,w1):
    return (2.1 - w0 - w1)**2 + (2.9 - w0 - 3*w1)**2 + (4.1 - w0 - 5*w1)**2

optimize.minimize(lambda x: f3(*x), (1,1))
```

```python
def f4(x,y,z):
    return x**3 - 2*x**2 + y**2 + z**2 - 2*x*y + x*z - y*z + 3*z

optimize.minimize(lambda x: f4(*x), (1,1,1))
```

```python
def f(x,y):
    return x**2 + 2*(1-y)**3

def grad(x,y):
    dx = 2*x
    dy = 2*y
    return np.array([dx, dy])

x0 = np.array([-1.5,2])
gamma = 0.1
i = 0
x_cur = x0
vals = []
coords = []

while True:
    vals.append(f(*x_cur))
    coords.append(x_cur)
    print(f"iter = {i}, (x,y) = {x_cur}, f(x,y) = {f(*x_cur):.3f}, grad = {grad(*x_cur)}")
    if f(*x_cur) < 0.001 or i > 5000:
        break
    i += 1
    x_new = x_cur - gamma*grad(*x_cur)
    x_cur = x_new
```

```python
n = 100 # number of discretization points along the x-axis
m = 100 # number of discretization points along the x-axis
a=-2; b=1.5 # extreme points in the x-axis
c=-0.5; d=2.5 # extreme points in the y-axis
X,Y = np.meshgrid(np.linspace(a,b,n), np.linspace(c,d,m))

Z = f(X,Y)

plt.xlim(a,b)
plt.ylim(c,d)

plt.contour(X,Y,Z, np.logspace(0, 1, 20),cmap='gray_r')
plt.plot([c[0] for c in coords], [c[1] for c in coords])
plt.scatter(0,0,c='red')
plt.show()
```

```python
# преообразование через функцию apply:
# df.column_name.apply(func)
# func может быть определена через def или через lambda
sample.Age = sample.Age.apply(lambda x:x**2)
sample
```

```python
def age_category(age):
    if age <23:
        return 'молодой'
    elif 23<=age<=35:
        return 'средний'
    elif age>35:
        return 'зрелый'
    else:
        return
        
sample['Age_category'] = sample.Age.apply(age_category)
sample
```

```python
log = pd.read_csv('log.csv', header=None)
log.columns = ['user_id','time','bet','win']

def convert_user_id(s):
    if s == "#error":
        s = ""
    else:
        s = str(s).replace("Запись пользователя № - ","")
    return s

def convert_time(s):
    if type(s) == str:
        s=str(s).replace("[","")
    return s

log.user_id = log.user_id.apply(convert_user_id)
log.time = log.time.apply(convert_time)

log.tail()
```

```python
log = pd.read_csv("log.csv",header=None)
log.columns = ['user_id', 'time', 'bet', 'win']

log.drop(log[log.user_id == "#error"].index, inplace=True)

def convert_user_id(s):
    return str(s).replace("Запись пользователя № - ","")
    
def convert_time(t):
    if type(t) == str:
        t = str(t).replace("[","")
    return t

log.user_id = log.user_id.apply(convert_user_id)
log.time = log.time.apply(convert_time)
log.tail()
```

```python
def exchange_rates(currency, format='full'):
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Server error.\nResponse status code: {response.status_code}.")
        return 0
    else:
        try:
            currencies = response.json()['Valute']
        except:
            print('Something go wrong...')
            return 0
        try:
            if format == 'full':
                return currencies[currency]
            elif format == 'value':
                return currencies[currency]['Value']
        except:
            print('No such currency')
            return 0
        
```

```python
exchange_rates('USD', format='value')
```

```python
def currency_name(ID):
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'
    response = requests.get(url)
    
    if response.status_code != 200:
        return response.status_code
    
    try:
        currencies = response.json()['Valute']
    except:
        print('JSON parser error')
        return
    
    id_curr_dict = dict(map(lambda x: (x['ID'], x['Name']), currencies.values()))
    
    try:
        result = id_curr_dict[ID]
    except:
        print('ID error')
        return
    
    return result
    
```

Напишите функцию get_smm_index(group_name, token), которая по имени группы и авторизационному токену API возвращает smm_index группы - сумму лайков, комментариев и репостов для последних 10 постов, поделённую на количество участников в группе.

```python
def get_smm_index(group_name, token):
    # количество участников в группе
    url = 'https://api.vk.com/method/groups.getMembers'
    params = {
    'group_id':group_name,
    'v':5.131,
    'access_token':token
    }
    members_count = requests.get(url, params=params).json()['response']['count']
    
    # сумма лайков, комментариев и репостов
    url = 'https://api.vk.com/method/wall.get'
    params = {
    'domain':group_name,
    'filter':'all',
    'count':10,
    'offset':0,
    'v':5.131,
    'access_token':token
    }

    data = requests.get(url, params=params).json()['response']['items']
    
    stats=0
    for record in data:
        stats += record['comments']['count'] + record['likes']['count'] + record['reposts']['count']
    
    return stats/members_count
```

```python
import requests
from datetime import datetime

TOKEN = 'bd609931bd609931bd609931e1bd1be35ebbd60bd609931df400a080084c6e2a88bf067'

vk_groups = [
    'tomsk123',
    'ghest_tomsk',
    'ya_tomich',
    'tomskie',
    'typical_tomsk',
    'tomskvkurse',
    'inctsk',
    'znaigorod',
    'goodcity70',
    'black_list_tomsk',
    'dortom'
]

def get_last_news(group_name, from_time=None, token=TOKEN):
    url = 'https://api.vk.com/method/wall.get'
    params = {
    'domain':group_name,
    'filter':'all',
    'count':10,
    'offset':0,
    'v':5.131,
    'access_token':token
    }

    data = requests.get(url, params=params).json()['response']['items']

    feed = {}
    for record in data:
        text = record['text']
        if text:
            date = datetime.utcfromtimestamp(record['date']+3600*7).strftime('%Y-%m-%d %H:%M:%S')
            link = 'https://vk.com/vk?w=wall'+str(record['owner_id'])+'_'+str(record['id'])
            feed[date] = [text, link]
    
    return feed
```

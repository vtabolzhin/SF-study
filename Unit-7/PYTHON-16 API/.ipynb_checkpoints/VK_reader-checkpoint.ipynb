```python
import requests
import pandas as pd
from datetime import datetime

url = 'https://api.vk.com/method/wall.get'
TOKEN = 'bd609931bd609931bd609931e1bd1be35ebbd60bd609931df400a080084c6e2a88bf067'

def get_last_news(group_name, day, time_start, time_end, url=url, token=TOKEN):

    feed = {}
    count = 10
    offset = 0
    keep_search = True
    
    print(f'–ß–∏—Ç–∞—é –Ω–æ–≤–æ—Å—Ç–∏ {group_name}...')
    # –ø—Ä–æ—Å–º–∞—Ç—Ä–∏–≤–∞–µ–º –ø–æ –æ–¥–Ω–æ–π –Ω–æ–≤–æ—Å—Ç–∏ –≤ —Ü–∏–∫–ª–µ –æ—Ç 18.00 –¥–æ 09.00
    while True:
        params = {
        'domain':group_name,
        'filter':'all',
        'count':count,
        'offset':offset,
        'v':5.131,
        'access_token':token
        }

        # —Å–º–æ—Ç—Ä–∏–º –ø–∞—á–∫—É –Ω–æ–≤–æ—Å—Ç–µ–π
        posts = requests.get(url, params=params).json()['response']['items']
        
        for post in posts:
            # –∏–∑ –Ω–∏—Ö –æ—Ç–±–∏—Ä–∞–µ–º —Ç–µ, —á—Ç–æ –≤—Ö–æ–¥—è—Ç –≤ –Ω—É–∂–Ω—ã–π –≤—Ä–µ–º–µ–Ω–Ω–æ–π –∏–Ω—Ç–µ—Ä–≤–∞–ª
            if time_start <= post['date'] <= time_end:
                text = post['text']
                if len(text) == 0:
                    text = '–¢–µ–∫—Å—Ç–∞ –Ω–µ—Ç'
                date = datetime.utcfromtimestamp(post['date']+3600*7).strftime('%Y-%m-%d %H:%M:%S')
                link = 'https://vk.com/vk?w=wall'+str(post['owner_id'])+'_'+str(post['id'])
                feed[date] = [text, link, group_name]
            # –ø—Ä–æ–≤–µ—Ä—è–µ–º, –µ—Å—Ç—å –ª–∏ –Ω–µ–∑–∞–∫—Ä–µ–ø–ª–µ–Ω–Ω—ã–µ –Ω–æ–≤–æ—Å—Ç–∏ —Ä–∞–Ω—å—à–µ –Ω–∏–∂–Ω–µ–π –≥—Ä–∞–Ω–∏—Ü—ã –∏–Ω—Ç–µ—Ä–≤–∞–ª–∞
            if post['date'] < time_start:
                try:
                    if post['is_pinned'] == 1:
                        keep_search = True
                except:
                    keep_search = False
                
        # –ø—Ä–æ–≤–µ—Ä–∫–∞
        if keep_search==True:
            offset += count
            continue
        else:
            print('–ü—Ä–æ—á–∏—Ç–∞–Ω–æ.')
            break

    return feed

vk_groups = [
    'tomsk123',
    'ghest_tomsk',
    'ya_tomich',
    'tomskie',
    'typical_tomsk',
    'tomskvkurse',
    'inctsk',
    'znaigorod',
    'goodcity70',
    'black_list_tomsk',
    'dortom'
]

day = input('–í–≤–µ–¥–∏—Ç–µ –¥–∞—Ç—É –≤ —Ñ–æ—Ä–º–∞—Ç–µ –î–î-–ú–ú-–ì–ì–ì–ì –∏–ª–∏ –Ω–∞–∂–º–∏—Ç–µ Enter –¥–ª—è —Ç–µ–∫—É—â–µ–≥–æ –¥–Ω—è:\n')

# –ø–æ–ø—Ä–æ–±—É–µ–º –ø—Ä–æ—á–∏—Ç–∞—Ç—å –Ω–µ–æ–±—Ö–æ–¥–∏–º—É—é –¥–∞—Ç—É
if len(day)>0:
    try:
        today = datetime.strptime(day, '%d-%m-%Y').date()
    except Exception as err:
        print('–û—à–∏–±–∫–∞: –Ω–µ–∫–æ—Ä—Ä–µ–∫—Ç–Ω–∞—è –¥–∞—Ç–∞')
        raise(err)
        
else:
    today = datetime.now()

print(f'–ß–∏—Ç–∞—é –Ω–æ–≤–æ—Å—Ç–∏ –∑–∞ {today}')
# –ë—É–¥–µ–º —á–∏—Ç–∞—Ç—å –Ω–æ–≤–æ—Å—Ç–∏ —Å 9.00 –¥–æ 18.00
time_start = datetime(today.year, today.month, today.day, 9, 0).timestamp()
time_end = datetime(today.year, today.month, today.day, 18, 0).timestamp()

try:
    df = {}
    for group in vk_groups:
        df.update(get_last_news(group_name=group, day=today, time_start=time_start, time_end=time_end))

    df = pd.DataFrame(df).T
    df.columns = ['text', 'link','group']
    print('–ó–∞–ø–∏—Å—ã–≤–∞—é —Ñ–∞–π–ª excel...')
    df.to_excel(today.strftime('%Y-%m-%d--%H-%M-%S')+'.xlsx')
    print('–ì–æ—Ç–æ–≤–æ, –º–æ–∂–µ—Ç–µ –ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è!')
except Exception as err:
    print('–ß—Ç–æ-—Ç–æ –ø–æ—à–ª–æ –Ω–µ —Ç–∞–∫',err)

input('–î–ª—è –≤—ã—Ö–æ–¥–∞ –Ω–∞–∂–º–∏—Ç–µ Enter.')
```

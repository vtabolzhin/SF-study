```python
def exchange_rates(currency, format='full'):
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'
    response = requests.get(url)
    if response.status_code != 200:
        print(f"Server error.\nResponse status code: {response.status_code}.")
        return 0
    else:
        try:
            currencies = response.json()['Valute']
        except:
            print('Something go wrong...')
            return 0
        try:
            if format == 'full':
                return currencies[currency]
            elif format == 'value':
                return currencies[currency]['Value']
        except:
            print('No such currency')
            return 0
        
```

```python
exchange_rates("USD")
```

```python
def currency_name(ID):
    url = 'https://www.cbr-xml-daily.ru/daily_json.js'
    response = requests.get(url)
    
    if response.status_code != 200:
        return response.status_code
    
    try:
        currencies = response.json()['Valute']
    except:
        print('JSON parser error')
        return
    
    id_curr_dict = dict(map(lambda x: (x['ID'], x['Name']), currencies.values()))
    
    try:
        result = id_curr_dict[ID]
    except:
        print('ID error')
        return
    
    return result
    
```

```python
currency_name('R01700J')
```

```python
# Импортируем библиотеки
from bs4 import BeautifulSoup
import requests

# Получаем данные
url = 'https://nplus1.ru/news/2019/06/04/slothbot'
response = requests.get(url)

# Создим объект BeautifulSoup, указывая html парсер
page = BeautifulSoup(response.text, 'html.parser')

# Все готово, чтобы получать данные из страницы
# Для начала получим title, отображающийся на закладках браузера
print(page.title)
page.title.text
```

```python
def wiki_header(url):
    response = requests.get(url)
    page = BeautifulSoup(response.text, 'html.parser')
    return page.find('h1').text
```

```python
wiki_header('https://en.wikipedia.org/wiki/Operating_system')
```

```python
url = 'https://www.kinopoisk.ru/film/1272246/'

def get_actors(url):
    response = requests.get(url)
    page = BeautifulSoup(response.text, 'html.parser')
    links = page.find_all('a', class_='styles_link__31CGv')
    return [link.text for link in links]
```

```python
get_actors('https://www.kinopoisk.ru/film/42326/')
```
